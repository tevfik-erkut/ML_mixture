It clearly shows that PPO algorithm has outperformed A2C. Why is that?

Because, PPO algorithm is actually an union of A2C (having multiple workers) and TRPO (it uses a trust region to improve the actor). The main idea is that after an update, the new policy should be not too far form the old policy. For that, PPO uses clipping to avoid too large update.
Therefore since our problem has 3 actional space, [x, x ,x] PPO has a better advantage!

For policies;

ActorCriticPolicy performs better than ActorCriticCNNPolicy in this problem. 
ActorCriticPolicy > Policy class for actor-critic algorithms (has both policy and value prediction)
CNN policy >  CNN policy class for actor-critic algorithms (has both policy and value prediction). Used by A2C, PPO and the likes.
