## Building Model

Let's try to predict monthly sales!

```python
# Creating a dataframe for monthly/yearly sales
df_month_year = df.groupby('Month_year').sum()['Sales'].reset_index()
df_month_year
```

Visualization of sales numbers in the past

```python
plt.figure(figsize=(15, 7))
plt.plot(df_month_year['Month_year'], df_month_year['Sales'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
df_forecast = df_month_year.set_index("Month_year")
df_forecast
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
X = df_forecast.values
size = int(len(X) * 0.8)
train, test = df_forecast[0:size], df_forecast[size:len(X)]
```

Lets plot decomposition!

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df_forecast, period=  12)
plt.figure(figsize = (15,15))
fig, axes = plt.subplots(4, 1, sharex=True)

result.observed.plot(ax = axes[0], color='r')
axes[0].set_ylabel('Observed')

result.trend.plot(ax = axes[1], color='g')
axes[1].set_ylabel('Trend')

result.seasonal.plot(ax = axes[2])
axes[2].set_ylabel('Seasonality')

result.resid.plot(ax = axes[3], color='k')
axes[3].set_ylabel('Residuals')
plt.tight_layout()
```

Looks like we have a upwards trend!

#### Sarima MODEL-1

```python
mod = SARIMAX(train,
            order=(1, 1, 5), # p d q
            seasonal_order=(1, 1, 1, 6),
            enforce_stationarity=False,
            enforce_invertibility=False)
results = mod.fit()

predictions = results.forecast(len(test))
predictions = pd.DataFrame(predictions).set_index(test.index).iloc[:,0]
plt.figure(figsize=(20,5))

plt.plot(df_forecast)
plt.plot(predictions)

plt.legend(('Data', 'Predictions'), fontsize=16)

plt.title('Predictions vs Data', fontsize=20)
plt.ylabel('Volume', fontsize=16)
```

### Sarima Model -2

```python
mod = SARIMAX(train,
            order=(5, 1, 1), # p d q
            seasonal_order=(4, 1, 1, 6),
            enforce_stationarity=False,
            enforce_invertibility=False)
results = mod.fit()

predictions2 = results.forecast(len(test))
predictions2 = pd.DataFrame(predictions).set_index(test.index).iloc[:,0]
plt.figure(figsize=(20,5))

plt.plot(df_forecast)
plt.plot(predictions2)

plt.legend(('Data', 'Predictions'), fontsize=16)

plt.title('Predictions vs Data', fontsize=20)
plt.ylabel('Volume', fontsize=16)
```

### Blending predictions

```python
prediction_blend = (predictions2*1.2 + predictions)/2.2
```

```python
plt.figure(figsize=(20,5))

plt.plot(df_forecast)
plt.plot(prediction_blend)

plt.legend(('Data', 'Predictions'), fontsize=16)

plt.title('Predictions vs Data', fontsize=20)
plt.ylabel('Volume', fontsize=16)
```

### PROPHET Model

```python
#%pip install prophet
from prophet import Prophet
```

```python
pjme_train = df_forecast[:35]
pjme_test = df_forecast[35:]
from prophet import Prophet
model = Prophet()
model.fit(pjme_train.reset_index() \
              .rename(columns={'Month_year':'ds',
                               'Sales':'y'}))
```

```python
pjme_train.head()
```

```python
# Prediction on training set with model
pjme_test_fcst = model.predict(df=pjme_test.reset_index() \
                                   .rename(columns={'Month_year':'ds', "Sales": "y"}))
```

```python
# Plotting the forecast
f, ax = plt.subplots(1)
f.set_figheight(5)
f.set_figwidth(15)
fig = model.plot(pjme_test_fcst,
                 ax=ax)
plt.show()
```

```python
pjme_test.index = pd.to_datetime(pjme_test.index)
```

```python
# Plot the forecast with the actuals
f, ax = plt.subplots(1)
f.set_figheight(5)
f.set_figwidth(15)
ax.scatter(pjme_test.index, pjme_test['Sales'], color='r')
fig = model.plot(pjme_test_fcst, ax=ax)
```

 Looking good!

I am choosing Prophet model to predict the future months sales. 

```python
future = model.make_future_dataframe(periods = 12, freq = "M")
forecast = model.predict(future)
```
